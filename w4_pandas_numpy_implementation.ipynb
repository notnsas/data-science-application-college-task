### **Mengimport Library `pd` & `np`**

```python
import numpy as np
import pandas as pd
```

### **Load Dataset**

```python
df = pd.read_csv('https://raw.githubusercontent.com/LatiefDataVisionary/data-science-application-college-task/refs/heads/main/datasets/superstore_sales_dataset.csv')
df.head()
```

### **Statistika Deskriptif**

```python
df.describe()
```

**Hitung rata-rata kolom `Sales`**

```python
rata_rata_penjualan = np.mean(df['Sales'])
print('Rata-Penjualan: {}'.format(round(rata_rata_penjualan, 2)))
```

**Hitung median kolom `Profit`**


```python
median_sales = np.median(df['Sales'])
print(f'Median Profit: {round(median_sales, 2)}')
```
